```python
base_map = folium.Map(
    location=[35.0, 105.0],
    tiles='http://{s}.basemaps.cartocdn.com/light_all/{z}/{x}/{y}{r}.png',
    attr="&copy; <a href='https://stadiamaps.com/' target='_blank'>Stadia Maps</a> &copy; <a href='https://openmaptiles.org/' target='_blank'>OpenMapTiles</a> &copy; <a href='https://www.openstreetmap.org/copyright' target='_blank'>OpenStreetMap</a>&copy; <a href='https://stamen.com/' target='_blank'>Stamen Design</a>",
    zoom_start=5
)

def clean_merge_with_buffer(gdf, buffer_distance=0.0001):
    """使用缓冲技术清理内部线条"""
    
    # 原始合并
    merged = unary_union(gdf.geometry)
    
    # 使用微小缓冲消除缝隙
    buffered = merged.buffer(buffer_distance)
    
    # 反向缓冲恢复原始尺寸
    cleaned = buffered.buffer(-buffer_distance)
    
    # 再次微小缓冲确保边界光滑
    final_cleaned = cleaned.buffer(buffer_distance/2).buffer(-buffer_distance/2)
    
    return final_cleaned
import os
from shapely.ops import unary_union
from shapely import make_valid
files = os.listdir('..\\..\\database\\上海市\\上海市')
for file in files:
    country = gpd.read_file(f'..\\..\\database\\上海市\\上海市\\{file}')
    country['geometry'] = country['geometry'].apply(
        lambda geom: make_valid(geom) if not geom.is_valid else geom
    )
    merged = clean_merge_with_buffer(country.geometry)
    
    folium.GeoJson(merged).add_to(base_map)
    merged = gpd.GeoDataFrame({'name': file[:-5],
                               'geometry':[merged]})
    
    merged.to_file(f'{file[:-5]}.geojson', driver='GeoJSON')
base_map.save("map_shanghai.html")
```
