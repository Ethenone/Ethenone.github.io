```python

import json
import pandas as pd
import folium
```

```python
airportdata = pd.read_csv('airports_data.csv',encoding='gb18030')
airportdata.head()
```

```python
airportgeo = {}
for i in range(len(airportdata)):
    airportgeo[airportdata['iata'][i]]=[airportdata['lat'][i],airportdata['lon'][i]]
print(airportgeo)   
```

```python
data = pd.read_excel('飞机乘坐记录.xlsx', sheet_name = 0)
data.head()
```

```python
m = folium.Map(location=airportgeo['SHA'],
               zoom_start=5,
               control_scale=True,
               control=False,
               tiles=None
              )

folium.TileLayer(tiles='http://{s}.basemaps.cartocdn.com/light_all/{z}/{x}/{y}{r}.png',
                 attr="&copy; <a href='https://stadiamaps.com/' target='_blank'>Stadia Maps</a> &copy; <a href='https://openmaptiles.org/' target='_blank'>OpenMapTiles</a> &copy; <a href='https://www.openstreetmap.org/copyright' target='_blank'>OpenStreetMap</a>&copy; <a href='https://stamen.com/' target='_blank'>Stamen Design</a>",
                 min_zoom=0,
                 max_zoom=19,
                 control=True,
                 show=True,
                 overlay=False,
                 name='transportation'
                ).add_to(m)

group_air = folium.FeatureGroup(name='air')
for i in range(len(data)):
    folium.PolyLine(
        locations = [
            airportgeo[data['start'][i]],airportgeo[data['depart'][i]]
        ],
        color='red',
        weight=3,
        opacity=0.3
    ).add_to(group_air)
    
    
    folium.Circle(
        location=airportgeo[data['start'][i]],
        radius=9000,   # 圆的半径
        popup=folium.Popup(data['起飞'][i],max_width=10),
        color='red', #'#FF1493',
        fill=True,
        fill_color='#FFD700'
    ).add_to(group_air)
    
    folium.Circle(
        location=airportgeo[data['depart'][i]],
        radius=9000,   # 圆的半径
        popup=folium.Popup(data['降落'][i],max_width=10),
        color='red', #'#FF1493',
        fill=True,
        fill_color='#FFD700'
    ).add_to(group_air)

group_air.add_to(m)
#m.save('map2.html')
```

```python
# with open('StationGeo.json','r',encoding='utf-8') as file:
#     statgeo = json.load(file)
# print(statgeo)
statdata = pd.read_csv('stations_data.csv',encoding='gbk')
statdata.head()
```

```python


statgeo = {}
for i in range(len(statdata)):
    lat = statdata['纬度'][i]
    lng = statdata['经度'][i]
    loc = [lat,lng]
#     loc = correct(loc)
    statgeo[statdata['车站'][i]]=loc
print(statgeo)    
```

```python
data2 = pd.read_excel('火车乘坐记录.xlsx', sheet_name = 0)
data2.head()
```

```python
#m = folium.Map(location=statgeo['上海'], zoom_start=5)

group_train = folium.FeatureGroup(name='train')
for i in range(len(data2)):
    folium.PolyLine(
        locations = [
            statgeo[data2['上车站'][i]],statgeo[data2['下车站'][i]]
        ],
        color='blue',
        weight=3,
        opacity=0.3
    ).add_to(group_train)
    folium.Circle(
        location=statgeo[data2['上车站'][i]],
        radius=9000,   # 圆的半径
        popup=folium.Popup(data2['上车站'][i],max_width=10),
        color='blue', #'#FF1493',
        fill=True,
        fill_color='#FFD700'
    ).add_to(group_train)
    folium.Circle(
        location=statgeo[data2['下车站'][i]],
        radius=9000,   # 圆的半径
        popup=folium.Popup(data2['下车站'][i],max_width=10),
        color='blue', #'#FF1493',
        fill=True,
        fill_color='#FFD700'
    ).add_to(group_train)

group_train.add_to(m)
folium.LayerControl(collapsed=False).add_to(m)
m.save('map_line.html')
```
