```python
import json
with open('activities.json','r') as file:
    data = json.load(file)
print(data)
```

```python
pl = data[0]['summary_polyline']
print(data[0]['summary_polyline'])
```

```python
import polyline
loc=polyline.decode(pl)
```

```python
import folium
 
# 创建一个地图对象
m = folium.Map(location=[22.89984, 1.6750], zoom_start=13)
 
# 添加折线
folium.PolyLine(
    locations=loc,
    color='red'
).add_to(m)
 
# 保存地图到HTML文件
m.save('map_with_polyline.html')
```
